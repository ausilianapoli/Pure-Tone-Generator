```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.io import wavfile
from scipy.io.wavfile import write

def pureTone(frequency = 440, duration = 5, samplerate = 44100):
    dt = 1/samplerate
    t = np.arange(0, duration, dt)
    tone = np.sin(2*math.pi*frequency*t)
    return tone
```

```python
samplerate = 44100
frequency = 1000
duration = 10

data = pureTone(frequency, duration, samplerate)
data2 = pureTone(frequency*2, duration, samplerate)
print(data.shape)
plt.plot(data[0:50])
plt.plot(data2[0:50])
plt.figure(2) #plotting the two pure tone
```

```python
data3 = data2 + data 
plt.plot(data3[0:50]) #plotting the sum of two pure tones
data4 = data3
data4[data4>1]=1
data4[data4<-1]=-1
plt.plot(data4[0:50]) #plotting other two pure tones with some manipulation
```

```python
#normalize data into range [-32768, 32767]
data = np.int16(data*(2**15-1))
data2 = np.int16(data2*(2**15-1))
data3 = np.int16((data3/np.max(abs(data3))*(2**15-1)))

#save generated pure tone
write('tone1000.wav', samplerate, data)
write('tone2000.wav', samplerate, data2)
write('toneMix.wav', samplerate, data3)
```
